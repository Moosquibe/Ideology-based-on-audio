# We can now go for all the lines in wordssyllablesformants.txt

This function reformats the data and for each tuple (speaker, word, vowel) produces a list ordered as follows:

$$[\text{frequence 1}, \text{frequence 2}, \text{duration}, \text{formant1}, \text{formant 2}]$$

where formant1 and formant2 are lists containing 3 frequences.

The function also populates a list of all the speakers.

```python
def reformat(l,vowels_dict,names_list):
    l=l.replace('[','')
    l=l.replace(']','')
    l=l.replace("'",'')
    l=l.replace(' ','')
    l=l.split(',')
    word=l[0]
    speaker=l[1].lower()
    for i in range(3,len(l),13):
        sublist=l[i:i+12]
        vowel=sublist[0]
        freq1=float(sublist[1])
        freq2=float(sublist[2])
        duration=float(sublist[3])
        formant1=[float(i) for i in sublist[4:7]]
        formant2=[float(i) for i in sublist[8:11]]
        if (speaker,word,vowel) in vowels_dict:
            v=vowels_dict[(speaker,word,vowel)]
            v[0]+=1
            v[1]+=freq1
            v[2]+=freq2
            v[3]+=duration
            for i in range(len(formant1)):
                v[4][i]+=formant1[i]
                v[5][i]+=formant2[i]

        else:
            vowels_dict[(speaker,word,vowel)]=[1,freq1,freq2,duration,formant1,formant2]    
    names_list.append(speaker)
```

```python
vowels_dict={}
names_list=[]
for line in list_formants:
    reformat(line,vowels_dict,names_list)
    
```
